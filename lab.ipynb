# 11주차 실습: 난장판 폴더 확장자별 자동 청소하기

이 실습 노트북은 파일 확장자를 기준으로 자동 폴더 생성 및 이동을 수행하는 파이썬 스크립트를 작성하는 공간입니다. **코드를 외워서 직접 짜지 마세요.** 아래 단계에 따라 AI(Gemini)에게 요구하고 코드를 가져와서 에러를 해결해 보세요.

## [1단계] 난장판(더미) 파일 준비하기
먼저, 청소할 가짜 쓰레기 파일들이 필요합니다. 코랩 환경 내에 임의의 파일 100개를 무작위로 생성하는 코드를 제미니에게 물어보고, 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "코랩 환경의 'downloads'라는 폴더 안에 임의의 가짜 파일 100개를 만들어주는 파이썬 코드를 짜줘. 확장자는 pdf, docx, jpg, png, txt, xlsx가 섞여 있게 해줘."

```python
# 여기에 제미니가 작성해 준 '더미 파일 생성 코드'를 붙여넣고 실행하세요.
import os
import random

# 1. 폴더 생성
folder_name = 'downloads'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
    print(f"'{folder_name}' 폴더를 생성했습니다.")

# 2. 설정 값 준비
extensions = ['pdf', 'docx', 'jpg', 'png', 'txt', 'xlsx']
num_files = 100

print(f"{num_files}개의 파일 생성을 시작합니다...")

# 3. 파일 생성 루프
for i in range(1, num_files + 1):
    # 임의의 확장자 선택
    ext = random.choice(extensions)

    # 파일명 설정 (예: file_001.pdf)
    file_name = f"file_{i:03d}.{ext}"
    file_path = os.path.join(folder_name, file_name)

    # 빈 파일 생성
    with open(file_path, 'w') as f:
        f.write(f"This is a dummy file: {file_name}")

print(f"✅ 성공: '{folder_name}' 폴더 안에 {num_files}개의 파일이 생성되었습니다.")
```

## [2단계] 확장자별 자동 분류 스크립트 작성
이제 위에서 만든 수많은 파일들을 확장자별로 새 폴더를 만들어 깔끔하게 이동시켜주는 분류 코드를 제미니에게 요청하고 아래 셀에 붙여넣어 실행하세요.

* 💡 **프롬프트 팁**: "'downloads' 폴더 안에 있는 파일들을 확장자별로 확인해서, 확장자 이름과 똑같은 새 폴더를 만들고 그 안으로 파일들을 모두 이동시켜주는 파이썬 코드를 만들어 줘. 이미 폴더가 있으면 만들지 말고 그냥 이동시켜 줘."

```python
# 여기에 제미니가 작성해 준 '파일 자동 분류 코드'를 붙여넣고 실행하세요.
import os
import shutil

# 1. 대상 폴더 설정
base_dir = 'downloads'

# 2. downloads 폴더 내의 파일 목록 가져오기
if os.path.exists(base_dir):
    file_list = os.listdir(base_dir)

    for file_name in file_list:
        # 파일의 전체 경로
        old_path = os.path.join(base_dir, file_name)

        # 디렉토리가 아닌 '파일'인 경우에만 실행
        if os.path.isfile(old_path):
            # 파일명에서 확장자 추출 (예: .pdf)
            # os.path.splitext는 ('파일명', '.확장자')의 튜플을 반환합니다.
            name, extension = os.path.splitext(file_name)

            # 확장자에서 마침표(.) 제거하고 대문자로 변환 (예: PDF, JPG)
            # 만약 확장자가 없는 파일이면 'ETC' 폴더로 분류
            ext_name = extension.strip('.').upper()
            if not ext_name:
                ext_name = 'ETC'

            # 3. 확장자 이름의 새 폴더 경로 설정
            new_folder_path = os.path.join(base_dir, ext_name)

            # 4. 폴더가 없으면 생성
            if not os.path.exists(new_folder_path):
                os.makedirs(new_folder_path)
                print(f"새 폴더 생성: {ext_name}")

            # 5. 파일 이동
            new_path = os.path.join(new_folder_path, file_name)
            shutil.move(old_path, new_path)
            print(f"이동 완료: {file_name} -> {ext_name}/")

    print("\n✅ 모든 파일 정리가 완료되었습니다.")
else:
    print(f"❌ '{base_dir}' 폴더를 찾을 수 없습니다.")
```
